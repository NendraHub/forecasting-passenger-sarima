# Pengertian Time Series

## Time Series
**Time series** adalah serangkaian data yang diamati atau direkam dalam urutan waktu tertentu. Data time series biasanya dikumpulkan pada interval waktu yang tetap, seperti harian, mingguan, bulanan, atau tahunan. Analisis time series bertujuan untuk memahami pola atau tren dalam data, meramalkan nilai di masa depan, dan mengidentifikasi anomali atau pola musiman. Time series memiliki karakteristik unik, seperti tren (arah umum data seiring waktu), musiman (pola berulang dalam periode waktu tertentu), siklus (fluktuasi dalam waktu yang lebih panjang), dan irregular atau noise (fluktuasi acak).

## Teori Sederhana
Time Series terdiri dari 4 komponen yakni **Tren, Musiman, Siklus dan Irregular atau Noise.**


*   Tren adalah Perubahan jangka panjang atau arah umum data. Contohnya adalah ketika penjualan meningkat dari tahun ke tahun, maka dapat dikatakan tren positif.
*   Musiman adalah pola berulang pada periode tertentu yang biasanya diakibatkan oleh faktor eksternal. Contohnya adalah penjualan es krim yang lebih tinggi pada musim panas.
*   Siklus adalah pola yang tidak tetap tetapi berulang yang terjadi dikarenakan periode yang lebih panjang daripada pola musiman. Contohnya adalah siklus ekonomi.
*   Irregular atau Noise adalah pola yang tidak dapat di prediksi yang tidak mengikuti pola tren, musiman, dan siklus





# Pengertian Model

## Model ARIMA
Model ARIMA (Autoregressive Integrated Moving Average) adalah model yang paling umum digunakan dalam analisis time series khususnya pada data time series stasioner. Model ini terdiri dari tiga komponen utama yakni AR (Autoregressive), I (Integrated) dan MA (Moving Average).

*   AR (Autoregressive) adalah  komponen yang menunjukkan bahwa nilai time series saat ini dipengaruhi oleh nilai-nilai sebelumnya. Artinya, model akan menggunakan nilai sebelumnya dari time series untuk membuat prediksi.
*   I (Integrated) adalah komponen yang menunjukkan bahwa model ARIMA dapat mengubah data non-stasioner menjadi stasioner dengan menggunakan proses differencing. Differencing dilakukan dengan menghitung selisih antara data pada waktu sekarang dan waktu sebelumnya, menghilangkan tren atau pola.
*   MA (Moving Average) adalah komponen yang menunjukkan bahwa prediksi time series saat ini dipengaruhi oleh nilai-nilai error dari lag sebelumnya.

Ketiga komponen tersebut dilambangkan dengan notasi (p, d, q) dimana p adalah orde AR, d adalah orde I, dan q adalah orde MA.






### Penjelasan Orde Notasi ARIMA

*   Orde AR (p) mengacu pada jumlah periode waktu sebelumnya yang digunakan untuk memprediksi nilai pada waktu tertentu. Jadi, p = 1 berarti menggunakan data satu periode sebelumnya untuk membuat prediksi.
*   Orde I (d)  mengacu pada jumlah kali kita melakukan differencing (pengurangan) untuk membuat data stasioner, yaitu menghilangkan tren atau pola yang tidak stabil. Misalnya, d = 1 berarti kita mengambil selisih antara data saat ini dan data sebelumnya untuk menghapus tren.
*   Orde MA (q) mengacu pada jumlah kesalahan atau error dari periode sebelumnya yang digunakan untuk memperbaiki prediksi. Jadi, q = 1 berarti kita menggunakan error dari satu periode sebelumnya untuk memperbaiki prediksi saat ini.






## Model SARIMA
Model SARIMA adalah perluasan dari model ARIMA yang menambahkan komponen musiman ke dalam model yang sangat berguna untuk data time series yang memiliki pola musiman. Model SARIMA memiliki notasi (p, d, q)(P, D, Q, s) dimana (p,d,q) adalah parameter yang digunakan pada ARIMA dalam bentuk non-musiman, sedangkan (P, D, Q, s) adalah parameter pada AR, I,dan MA dalam bentuk musiman. Sementara itu s adalah parameter musim yang digunakan.

### Penjelasan Orde Notasi SARIMA

*   Orde AR Musiman (P) adalah jumlah periode musiman sebelumnya yang digunakan untuk memprediksi nilai saat ini berdasarkan pola musiman.
*   Orde I Musiman (D) adalah berapa kali differencing musiman diperlukan untuk membuat data musiman menjadi stasioner.
*   Orde Q Musiman (Q) adalah jumlah kesalahan atau error musiman sebelumnya yang digunakan untuk memperbaiki prediksi berdasarkan error dari periode musiman yang lalu.
*   Parameter Musim (s) Menunjukkan panjang periode musiman. Misalnya, s = 12 berarti data memiliki pola musiman tahunan dengan 12 periode (seperti data bulanan dengan pola tahunan).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

## Membaca Dataset

```python
data = pd.read_csv('/kaggle/input/air-passengers/AirPassengers.csv')
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)
```

## Sample Dataset

```python
data.head()
```

## Melihat Tren dan Pola Musiman

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 4))
plt.plot(data['#Passengers'], color='blue', label='Jumlah Penumpang')
plt.title('Jumlah Penumpang Pesawat Per Bulan')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Penumpang')
plt.legend()
plt.show()
```

## Melakukan dekomposisi data

```python
decomposition = seasonal_decompose(data['#Passengers'], model='multiplicative')
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(6, 9))
decomposition.observed.plot(ax=ax1, legend=False)
ax1.set_ylabel('Data Asli')
decomposition.trend.plot(ax=ax2, legend=False)
ax2.set_ylabel('Trend')
decomposition.seasonal.plot(ax=ax3, legend=False)
ax3.set_ylabel('Seasonal')
decomposition.resid.plot(ax=ax4, legend=False)
ax4.set_ylabel('Noise')
plt.tight_layout(pad=2)
plt.show()
```

Pada grafik seasonal dan noise (residual) dalam model multiplikatif, sumbu y dengan nilai seperti 0.8, 1.0, dan 1.2 menunjukkan rasio atau proporsi relatif terhadap data asli. Nilai ini menggambarkan variasi musiman atau fluktuasi acak dalam bentuk persentase. Misalnya, nilai 1.2 pada seasonal berarti 20% di atas rata-rata, sementara 0.8 berarti 20% di bawah rata-rata. Ini mencerminkan seberapa besar komponen seasonal atau noise menyimpang dari nilai dasar data.

## Menentukan Rentang Data Train dan Test

```python
# Tentukan train dan test
train = data[:'1959']
test = data['1960':]
```

Data train mencakup seluruh data dari tahun 1949 hingga Desember 1959 dan data test dimulai dari Januari 1960.

## Melatih Model

```python
warnings.filterwarnings("ignore")

model = SARIMAX(train['#Passengers'],
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 12),
                enforce_stationarity=False,
                enforce_invertibility=False)
sarima_model = model.fit(disp=False)
```

Parameter order=(1, 1, 1) menentukan model ARIMA untuk data non-musiman, sementara seasonal_order=(1, 1, 1, 12) menentukan model musiman dengan panjang musim 12 (pola tahunan).

## Ringkasan Model

```python
print(sarima_model.summary())
```

## Prediksi Berdasarkan Data Test

```python
future_pred = sarima_model.get_forecast(steps=12)
future_ci = future_pred.conf_int()
```

## Menghitung MSE dan RMSE

```python
pred = future_pred
mse = mean_squared_error(test['#Passengers'], pred.predicted_mean)
rmse = np.sqrt(mse)
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

MSE Senilai 467 menunjukkan bahwa rata-rata kuadrat kesalahan antara prediksi dan data aktual cukup besar. Artinya, prediksi model tidak terlalu akurat jika dibandingkan dengan nilai sebenarnya. Sedangkan, RMSE senilai 21.62 menunjukkan bawah secara rata-rata prediksi model berbeda 21.62 unit dari nilai aktual (Berbeda 21.62 penumpang per bulan).

## Visualisasi Plot Hasil Prediksi

```python
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['#Passengers'], label='Data Latih')
plt.plot(test.index, test['#Passengers'], color='orange', label='Data Aktual')
plt.plot(test.index, pred.predicted_mean, color='green', label='Prediksi')
plt.fill_between(test.index, future_ci.iloc[:, 0], future_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Prediksi Jumlah Penumpang dengan SARIMA')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Penumpang')
plt.legend()
plt.show()
```

## Perbedaan Secara Nominal

```python
predicted_values = pred.predicted_mean
actual_values = test['#Passengers']
differences = actual_values - predicted_values

difference_df = pd.DataFrame({
    'Actual': actual_values,
    'Predicted': predicted_values,
    'Difference': differences
})

difference_df
```

Berdasarkan hasil perbandingan antara data aktual dan prediksi model SARIMA, terlihat bahwa selisih antara keduanya bervariasi dengan beberapa bulan menunjukkan prediksi yang sangat dekat dengan data aktual (misalnya Juni dan Juli 1960) dan beberapa bulan lainnya menunjukkan selisih yang cukup besar (misalnya Maret dan November 1960). Meskipun ada beberapa deviasi, perbedaan rata-rata yang relatif kecil menunjukkan bahwa model memiliki akurasi yang cukup baik dalam memprediksi jumlah penumpang pesawat,
